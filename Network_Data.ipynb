```python
import pandas as pd
from pandas import DataFrame
```

```python
data = pd.read_excel('COVID19UGRecovery.xlsx',index=False)
data.head()
```

```python
data_edge=data[['Username','Retweeted By']]
data_edge = data_edge.rename(columns={'Username': 'Source','Retweeted By':'Target'})
#data_edge.columns=['Username','Retweeted_By']
#data_edge1=data_edge[data_edge.RetweetedBy!='']
data_edge1=data_edge.dropna()
data_edge1.head()
```

```python
output=data_edge1.to_excel('COVID19UGRecovery_edge.xlsx')
```

```python
source=data_edge1[['Source']]
source = source.rename(columns={'Source': 'Name'})
target=data_edge1[['Target']]
target = target.rename(columns={'Target': 'Name'})
```
